## Observation

```python
ratings = df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
ratings
```

```python
plt.rcParams["figure.figsize"] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",hue="Rating color",palette=["purple","red","orange","yellow","green","green"],data=ratings)
```

## Observation
1. Not rated count is very high.
2. Maximum Observation are between 2.5 and 3.4


```python
sns.countplot(x="Rating color",data=ratings)
```
